```python
class Game():
    def __init__(self, init_cards_1, init_cards_2):
        self.p1 = Queue()
        self.p2 = Queue()
        for card in init_cards_1:
            self.p1.put(card)
        for card in init_cards_2:
            self.p2.put(card)
        self.tot_cards = self.p1.qsize() + self.p2.qsize()
        self.num_rounds = 1
        self.keep_playing = True
        self.winner = None
        
    def checkGameState(self):
        if self.p1.empty():
            self.keep_playing = False
            self.winner = self.p2
            print('')
            print('Player 2 Won!')
            print(f"Player 2's deck: {list(self.winner.queue)}")
            self.computeScore()
            print(f"Final score: {self.score}")
        elif self.p2.empty():
            self.keep_playing = False
            self.winner = self.p1
            print('')
            print('Player 1 Won!')
            print(f"Player 1's deck: {list(self.winner.queue)}")
            self.computeScore()
            print(f"Final score: {self.score}")
        
    def playRound(self):
        print(f'-- Round {self.num_rounds} --')
        print(f"Player 1's deck: {list(self.p1.queue)}")
        print(f"Player 2's deck: {list(self.p2.queue)}")
        self.card1 = self.p1.get()
        self.card2 = self.p2.get()
        print(f"Player 1's playes: {self.card1}")
        print(f"Player 2's playes: {self.card2}")
        if self.card1 > self.card2:
            self.p1.put(self.card1)
            self.p1.put(self.card2)
            print('Player 1 wins the round!')
        else:
            self.p2.put(self.card2)
            self.p2.put(self.card1)
            print('Player 2 wins the round!')
        self.num_rounds += 1
        self.checkGameState()
    
    def computeScore(self):
        self.score = 0
        for i in range(self.tot_cards):
            self.score += self.winner.get() * (self.tot_cards - i)
```

```python
class Game():
    def __init__(self, init_cards_1, init_cards_2, n_game):
        self.p1 = Queue()
        self.p2 = Queue()
        for card in init_cards_1:
            self.p1.put(card)
        for card in init_cards_2:
            self.p2.put(card)
        self.tot_cards = self.p1.qsize() + self.p2.qsize()
        self.n_cards_1 = self.p1.qsize()
        self.n_cards_2 = self.p2.qsize()
        self.num_rounds = 1
        self.keep_playing = True
        self.winner = None
        self.old_rounds_1 = [list(self.p1.queue)]
        self.old_rounds_2 = [list(self.p2.queue)]
        self.n_game = n_game
    
    def playRound(self):
        print(f'=== Game {self.n_game} ===')
        self.p1_card = self.p1.get()
        self.p2_card = self.p2.get()
        
        # check size
        self.n_cards_1 = self.p1.qsize()
        self.n_cards_2 = self.p2.qsize()
        
        # check remaining
        if self.n_cards_1 >= self.p1_card and self.n_cards_2 >= self.p2.card:
            # create subgame!
            deck_1 = []
            deck_2 = []
            for i in range(self.n_cards_1):
                deck_1.append(self.p1.get())
            for i in range(self.n_cards_2):
                deck_2.append(self.p2.get())
                
            subg = SubGame(deck_1, deck_2, self.n_game)
            while subg.keep_playing:
                subg.playRound()
            self.winner = subg.winner
            
        
        else:
            if self.p1_card > self.p2_card:
                print(f'Player 1 wins round {self.num_rounds} of game {self.n_game}!')
            else:
                print(f'Player 2 wins round {self.num_rounds} of game {self.n_game}!')
                
        self.num_rounds += 1 
        self.n_game += 1
    
            
        
        
```
