```python
class Grid():
    def __init__(self, initial_grid):
        self.grid = {}
        self.states = {}
        for i in range(5):
            for j in range(5):
                self.grid[(i, j)] = initial_grid[i][j]
                self.states[(i, j)] = initial_grid[i][j]
                
        self.values = list(self.grid.values())
        self.steps = 0
        self.previous_layouts = [self.states.copy()]
        self.run = True
    
    def printGrid(self):
        i = 0
        if self.steps == 0:
            print('Initial State:')
        elif self.steps == 1:
            print('After 1 minute:')
        else:
            print(f'After {self.steps} minutes:')
        while i < len(self.values):
            line = []
            for j in range(5):
                line.append(self.values[i])
                i += 1
            print(''.join(line))
        print()
        
    def updateState(self):
        for i in range(5):
            for j in range(5):
                tot_bugs = 0
                v_movements = [i-1, i+1]
                h_movements = [j-1, j+1]
                for movement in h_movements:
                    try:
                        if self.grid[(i, movement)] == '#':
                            tot_bugs += 1
                    except:
                        continue
                for movement in v_movements:
                    try: 
                        if self.grid[(movement, j)] == '#':
                            tot_bugs += 1
                    except:
                        continue
                if tot_bugs != 1 and self.grid[(i, j)] == '#':
                    self.states[(i,j)] = '.'
                if self.grid[(i,j)] == '.' and tot_bugs in [1, 2]:
                    self.states[(i,j)] = '#'
        self.previous_layouts.append(self.states.copy())

    def updateGrid(self):
        self.updateState()
        self.grid = self.states.copy()
        self.values = list(self.grid.values())
        self.steps += 1
        
    def checkLayouts(self):
        for i in range(len(self.previous_layouts) - 1):
            if self.states == self.previous_layouts[i]:
                self.run = False
                print(f' --- Layout already occurred! --- ')
                self.printGrid()
                
    def computeRating(self):
        self.tot = 0
        for i in range(len(self.values)):
            if self.values[i] == '#':
                self.tot += 2**i
        print(f'Rating: {self.tot}')
        
    def runSimulation(self):
        self.printGrid()
        while self.run:
            self.updateGrid()
            self.printGrid()
            self.checkLayouts()
        self.computeRating()
```

```python
init_input = ['..##.',
        '#....', 
        '.....',
        '#.#.#',
        '#..#.']

g = Grid(init_input)
g.runSimulation()
```
